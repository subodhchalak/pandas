```python
# Anscombe's quartet

import seaborn as sns
anscombe = sns.load_dataset('anscombe')
print(anscombe)
print(type(anscombe))
```

```python
# matplotlib basics
import matplotlib.pyplot as plt

# create a subset of the data
# contains only data set 1 from anscombe
data_set_1 = anscombe[anscombe['dataset'] == 'I']
print(data_set_1)

# add x and y column to plot the graph
plt.plot(data_set_1['x'], data_set_1['y'])
plt.title("Anscombe's Quartet - Dataset I")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

# if we want to show only point, we can user 'o' param
sns.barplot(x='x', y='y', data=data_set_1)
plt.title("Seaborn Scatter - Dataset I")
plt.show()
```

```python
# figure objects and axes subplots

# create a entire figure where our subplots will go
fig = plt.figure()

data_set_1 = anscombe[anscombe['dataset'] == 'I']
data_set_2 = anscombe[anscombe['dataset'] == 'II']
data_set_3 = anscombe[anscombe['dataset'] == 'III']
data_set_4 = anscombe[anscombe['dataset'] == 'IV']

axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)

axes1.plot(data_set_1['x'], data_set_1['y'], 'o')
axes2.plot(data_set_2['x'], data_set_2['y'], 'go')
axes3.plot(data_set_3['x'], data_set_3['y'], 'ro')
axes4.plot(data_set_4['x'], data_set_4['y'], 'bo')

# add title to figure
fig.suptitle("Anscombe Data")
fig.tight_layout()

plt.show()
```

```python
# statistical graphics using matplotllib

tips = sns.load_dataset('tips')
print(tips)
# print(tips['total_bill'].max())
```

```python
# create a univariate figure
fig = plt.figure()

# create subplot with 1 row, 1 column and plot location 1
axes1 = fig.add_subplot(1, 1, 1)
axes1.hist(data=tips, x='total_bill', bins=10)
plt.title('Total Bill Distribution')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.histplot(data=tips, x="total_bill", kde=True)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title("Histogram with KDE")
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
plt.title("Total Bill vs Tip by Sex, Time and Size")
plt.show()
```
