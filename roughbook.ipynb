```python
def gen_func(n):
    i = 0
    while i < n:
        yield i
        i += 1

mygen = gen_func(5)

for num in mygen:
    print(num)
```

```python
from multiprocessing import Process
import time

def task1():
    print("Task1 function is running....")
    time.sleep(4)
    print("Task1 completed...")

def task2():
    print("Task2 function is running....")
    time.sleep(4)
    print("Task2 completed...")

p1 = Process(target=task1)
p2 = Process(target=task2)
p1.start()
# p1.join()
print("\n")
p2.start()
```

```python
from threading import Thread
import time

def task1(a, b):
    print("task1 started...")
    time.sleep(5)
    print("task1 completed...", a, b)


def task2():
    print("task2 started...")
    time.sleep(5)
    print("task2 completed...")


t1 = Thread(target=task1, kwargs={'a': 5, 'b': 10})
t2 = Thread(target=task2)

t1.start()
t1.join()
t2.start()
```

```python
def decorator_func(func):
    def wrapper(*args, **kwargs):
        print("This is decorator func")
        a, b = args
        if b <= 0:
            return "Error: denonminator should be greater than 0"
            
        result = func(*args, **kwargs)
        return result
        print("decorator ended")

    return wrapper

@decorator_func
def division(a, b):
    print(f"Hello from greet function: {a}, {b}")
    return a/b


final_result = division(10, 0)
print(final_result)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = sns.load_dataset('tips')

# Plot histogram of total_bill
dataset['total_bill'].plot.hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.show()
```

```python
dataset[['total_bill', 'tip']].plot.hist(alpha=0.8, bins=20)
plt.title('Total Bill vs Tip Distribution')
plt.show()
```

```python
plt.hist(dataset['total_bill'], bins=20, color='orange', edgecolor='black')
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist2d(dataset['total_bill'], dataset['tip'], bins=20, cmap='Blues')
plt.colorbar(label='Counts')
plt.title('2D Histogram: Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.histplot(data=dataset, x='total_bill', bins=20, kde=False)
plt.title('Histogram of Total Bill')
plt.show()
```

```python
sns.histplot(data=dataset, x='total_bill', hue='sex', bins=20, kde=True, multiple='stack')
plt.title('Total Bill Distribution by Sex')
plt.show()
```
